```python
import networkx as nx
import math

def calc_dist(a, b, features=2):
    sum = 0;
    for i in range(features):
        sum += (a[i] - b[i]) ** 2;
    return math.sqrt(sum)

k: int = centers

G = nx.DiGraph()

G.add_node('s', demand=-k)
G.add_node('t', demand=k)

num_of_nodes = len(list_x)
print(num_of_nodes)
for i in range(num_of_nodes):
    G.add_node(f'{i}.s', demand=1)
    G.add_node(f'{i}.t', demand=-1)


for i in range(num_of_nodes):
    for j in range(i+1, num_of_nodes):     
        G.add_edge(f'{i}.t', f'{j}.s', capacity=1, weight=calc_dist(list_x[i], list_x[j], features))
        
for i in range(num_of_nodes):
    G.add_edge('s', f'{i}.s', capacity=1, weight=0)
    G.add_edge(f'{i}.t', 't', capacity=1, weight=0)

flow_cost, flow_dict = nx.capacity_scaling(G)

print(flow_cost)
for key, value in flow_dict.items():
    print(key, value)
```

# DBScan Algorithm

The network is designed to find core nodes and then we use the answer to construct the solution. <br>
The network consists of two rows of n nodes with an additional 2 nodes for each data point, and a dummy node to balance the network <br>

Let's call the first row correspondent of each data point i, i.1 and each second row correspondent i.2.
Each first row node is connected to each second row node with capacity of 1 if the distance between them was less than epsilon and capacity of 0 otherwise. the weight is -1. <br>

For each i, i.2 is connected to i-d with capacity minpt-1 and weight of 0. <br>
For each i, i.2 is also connected to i-c with wieght of 1. (ensuring that i-d is used up to all it's capacity before flow goes into i-c ensring that flow goes to i-c iff more than minpt flow goes into i.2) <br>

Each first row node starts with $n$ flow.(demand = n) <br>
The dummy node is used to balance the network. Each first row node is connected to the dummy node with weight of 1. Each i-c and i-d is also connected to the dummy node with weight 0. <br>
The dummy node has $n^2$ outward flow. (demand = $-n^2$). <br>

```python
import networkx as nx
import math

def calc_dist(a, b, features=2):
    sum = 0;
    for i in range(features):
        sum += (a[i] - b[i]) ** 2;
    return math.sqrt(sum)


def solve_DBScan(minpt, epsilon):
    DBScan_G = nx.DiGraph()
    num_of_nodes = len(list_x)

    DBScan_G.add_node('dummy', demand=num_of_nodes**2)

    for i in range(num_of_nodes):
        DBScan_G.add_node(f'{i}.1', demand=-num_of_nodes)
        DBScan_G.add_node(f'{i}.2')
        DBScan_G.add_edge(f'{i}.1', 'dummy', weight=1)

    for i in range(num_of_nodes):
        for j in range(i+1, num_of_nodes):
            DBScan_G.add_edge(f'{i}.1', f'{j}.2', capacity=1 if int(epsilon/calc_dist(list_x[i] , list_x[j], features)) >= 1 else 0, weight=-1) 
            DBScan_G.add_edge(f'{j}.1', f'{i}.2', capacity=1 if int(epsilon/calc_dist(list_x[i] , list_x[j], features)) >= 1 else 0, weight=-1) 

    for i in range(num_of_nodes):
        DBScan_G.add_edge(f'{i}.2', f'{i}-d', capacity=minpt-1, weight=0)
        DBScan_G.add_edge(f'{i}.2', f'{i}-c', capacity=num_of_nodes, weight=1)

        DBScan_G.add_edge(f'{i}-d', 'dummy', weight=0)
        DBScan_G.add_edge(f'{i}-c', 'dummy', weight=0)


    DB_flow_cost, DB_flow_dict = nx.capacity_scaling(DBScan_G)

    cores = [k[:-2] for key, value in DB_flow_dict.items() for k, v in DB_flow_dict[key].items() if k[-2:]=='-c' and v > 0]
    edges = [(int(k[:-2]), int(key[:-2])) for key, value in DB_flow_dict.items() for k, v in DB_flow_dict[key].items() if k[:-2] in cores and v == 1]

    Gg = nx.Graph(edges)  
    components = list(nx.connected_components(Gg))  # Find connected components

    clusters = [0 for _ in range(num_of_nodes)]
    i = 1
    for component in components:
        tmp = list(component)
        for x in tmp:
            clusters[x] = i
        i += 1
        
    return (list_x, clusters)
```

```python
def calc_DBI(nodes, clusters):
    num_of_clusters = len(set(clusters))
    
    
    cluster_sum = [[0, 0] for _ in range(num_of_clusters)]
    cluster_nodes = [0 for _ in range(num_of_clusters)]
    centroids = [[0 , 0] for _ in range(num_of_clusters)]
    
    
    for i in range(len(clusters)):
        cluster_sum[clusters[i]][0] += nodes[i][0]
        cluster_sum[clusters[i]][1] += nodes[i][1]
        cluster_nodes[clusters[i]] += 1
    

    for i in range(num_of_clusters):
        centroids[i][0] = cluster_sum[i][0] / cluster_nodes[i]
        centroids[i][1] = cluster_sum[i][1] / cluster_nodes[i]
    
    
    compactness = [0 for _ in range(num_of_clusters)]
    for i in range(len(clusters)):
        compactness[clusters[i]] += calc_dist(centroids[clusters[i]], nodes[i])

    for i in range(num_of_clusters):
        compactness[i] = compactness[i] / cluster_nodes[i]
        
    
    seperation = [[0 for _ in range(num_of_clusters)] for _ in range(num_of_clusters)]
    for i in range(num_of_clusters):
        for j in range(i+1, num_of_clusters):
            seperation[i][j] = calc_dist(centroids[i], centroids[j])
            seperation[j][i] = calc_dist(centroids[i], centroids[j])
    
    
    R = [[0 for _ in range(num_of_clusters)] for _ in range(num_of_clusters)]
    for i in range(num_of_clusters):
        for j in range(i+1, num_of_clusters):
            R[i][j] = (compactness[i] + compactness[j]) / seperation[i][j]
            R[j][i] = (compactness[i] + compactness[j]) / seperation[i][j]
    for i in range(num_of_clusters):
        R[i][i] = 100000
    
    R = [min(R[i]) for i in range(num_of_clusters)]
    
    return sum(R) / num_of_clusters
```

```python
epsilon_range = 25 + 1
minpt_range = 10 + 1
epsilon_list = [0.01 * i for i in range(1, epsilon_range)]
minpt_list = [i for i in range(1, minpt_range)]

best = 0
best_DBI = 100000000
best_index = (0, 0)
for i in range(len(minpt_list)):
    for j in range(len(epsilon_list)):
        print(i, j)
        nodes, clusters = solve_DBScan(minpt_list[i], epsilon_list[j])
        tmp = clusters
        if not 0 in clusters:
            clusters = [clusters[i]-1 for i in range(len(clusters))]

        tmp_DBI = calc_DBI(nodes, clusters)
        if(tmp_DBI < best_DBI):
            best = tmp
            best_DBI = tmp_DBI
            best_index = (i, j)

print(best)
print(i+1, j*.01)
```
